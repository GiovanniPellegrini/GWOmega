```python
#create grid for RD
k_peak_values = np.geomspace(2*np.pi, 300*2*np.pi, 75)
sigma_values = np.linspace(0.1, 1, 75)
df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values
omega_grid_RD=compute_integral_grid_RD(k_values, k_peak_values, sigma_values)
save_grid_RD("omega_grid_RD.pkl", k_peak_values, sigma_values, omega_grid_RD)
```

```python
# create grid for eMD
k_max_values = np.geomspace(2*np.pi, 300*2*np.pi, 150)
x_R_values=np.geomspace(50, 450, 150)
df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values

omega_grid = compute_integral_grid_eMD(k_values, k_max_values, x_R_values)
save_grid_eMD("omega_grid_eMD_v7.pkl", k_max_values, x_R_values, omega_grid)
```

```python
#create different grid with different densities

k_max_values_1=np.linspace(1*2*np.pi, 50*2*np.pi, 300)
k_max_values_2=np.linspace(50*2*np.pi, 300*2*np.pi, 30)
k_max_values=np.unique(np.concatenate((k_max_values_1, k_max_values_2)))
x_R_values=np.geomspace(50, 450, 150)

df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values

omega_grid = compute_integral_grid_eMD(k_values, k_max_values, x_R_values, checkpoint_file="emd_grid_checkpoint.pkl")
save_grid_eMD("omega_grid_eMD_v9.pkl", k_max_values, x_R_values, omega_grid)

```

```python
import pickle
import numpy as np

with open('omega_grid_eMD_v8.pkl', 'rb') as f:
    data = pickle.load(f)

k_max_values = data['k_max_values']
x_R_values   = data['x_R_values']
omega_grid   = data['Omega_grid']

# Trova e rimuovi il duplicato
_, unique_idx = np.unique(k_max_values, return_index=True)
print(f"Punti totali: {len(k_max_values)}, unici: {len(unique_idx)}")
print(f"Duplicati rimossi: {len(k_max_values) - len(unique_idx)}")

k_max_clean  = k_max_values[unique_idx]
omega_clean  = omega_grid[unique_idx, :, :]

# Salva il file corretto
data['k_max_values'] = k_max_clean
data['Omega_grid']   = omega_clean

with open('omega_grid_eMD_v8.pkl', 'wb') as f:
    pickle.dump(data, f)

print("Fatto! Nuova shape:", omega_clean.shape)
```
