```python
#create grid for RD
k_peak_values = np.geomspace(2*np.pi, 300*2*np.pi, 75)
sigma_values = np.linspace(0.1, 1, 75)
df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values
omega_grid_RD=compute_integral_grid_RD(k_values, k_peak_values, sigma_values)
save_grid_RD("omega_grid_RD.pkl", k_peak_values, sigma_values, omega_grid_RD)
```

```python
# create grid for eMD
k_max_values = np.geomspace(2*np.pi, 300*2*np.pi, 150)
x_R_values=np.geomspace(50, 450, 150)
df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values

omega_grid = compute_integral_grid_eMD(k_values, k_max_values, x_R_values)
save_grid_eMD("omega_grid_eMD_v7.pkl", k_max_values, x_R_values, omega_grid)
```

```python
#create different grid with different densities

k_max_values_1=np.linspace(1, 50, 300)*2*np.pi
k_max_values_2=np.linspace(50, 300, 30)*2*np.pi
k_max_values=np.concatenate((k_max_values_1, k_max_values_2))
x_R_values=np.geomspace(50, 450, 150)

df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values

omega_grid = compute_integral_grid_eMD(k_values, k_max_values, x_R_values)
save_grid_eMD("omega_grid_eMD_v9.pkl", k_max_values, x_R_values, omega_grid)

```
