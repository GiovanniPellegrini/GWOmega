```python
#create grid for RD
k_peak_values = np.linspace(2*np.pi, 300*2*np.pi, 100)
sigma_values = np.logspace(-2, 0, 100)
df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values
omega_grid_RD=compute_integral_grid_RD(k_values, k_peak_values, sigma_values)
save_grid_RD("omega_grid_RD.pkl", k_peak_values, sigma_values, omega_grid_RD)
```

```python
# create grid for eMD
k_max_values = np.linspace(2*np.pi, 300*2*np.pi, 100)
eta_R_values = np.linspace(120/k_max_values[-1], 120/k_max_values[0], 100) 
df=1/10
f_values=np.arange(1, 200+df, df)
k_values=2*np.pi*f_values

omega_grid = compute_integral_grid_eMD(k_values, k_max_values, eta_R_values)
save_grid_eMD("omega_grid_eMD_75x75.pkl", k_max_values, eta_R_values, omega_grid)
```
